# Calculation of Wave Activity Flux (WAF) using python

This python code is based on the grads code (http://www.atmos.rcast.u-tokyo.ac.jp/nishii/programs/tnfxy-grads.ncep.nov2014.txt) that calculates the horizontal wave-activity flux derived by Takaya and Nakamura (1999, 2001).

### Import python packages

```python
import sys
print(sys.path)

import xarray as xr
import metpy.calc as mpcalc
from metpy.units import units
import numpy as np
import matplotlib.pyplot as plt

from scipy.constants import pi, g, R
```

### Download data

Used data: monthly mean data of NCEP/NCAR reanalys 1

The data are available at http://www.esrl.noaa.gov/psd/data/gridded/data.ncep.reanalysis.html

```python
from urllib import request
from os.path import isfile

url = 'ftp://ftp.cdc.noaa.gov/Datasets/ncep.reanalysis.derived/pressure/'
for fname in ['uwnd.mon.mean.nc', 'vwnd.mon.mean.nc', 'hgt.mon.mean.nc']:
    if isfile(fname):
        print(" skip to download ... "+fname)
    else:
        print(" downloading ... "+fname)
        request.urlretrieve(url+fname, fname)
```

### Read netcdf data using xarray

Variables: Monthly-mean zonal and meridional wind (uwnd, vwnd), geopotential height (hgt) at 250 hPa.

```python
plev = 250
year = 2014
month = 11 # November
yyyymm = "{:04d}-{:02d}".format(year, month)

ds = xr.open_dataset("uwnd.mon.mean.nc").metpy.parse_cf()
uwnd = ds.uwnd.sel(time=ds.time.dt.month.isin([month]), 
                   level=plev).reindex(lat=list(reversed(ds.lat)))

ds = xr.open_dataset("vwnd.mon.mean.nc").metpy.parse_cf()
vwnd = ds.vwnd.sel(time=ds.time.dt.month.isin([month]), 
                   level=plev).reindex(lat=list(reversed(ds.lat)))

ds = xr.open_dataset("hgt.mon.mean.nc").metpy.parse_cf()
hgt = ds.hgt.sel(time=ds.time.dt.month.isin([month]), 
                   level=plev).reindex(lat=list(reversed(ds.lat)))

hgt.sel(time=yyyymm).squeeze().plot.contourf()
```

Basic state: climatology of monthly-mean field (January) averaged from 1980 to 2000.

```python
# Calculate monthly climatology for 1980-2000 years.

start_date, end_date = ["1980-01", "2000-12"]
uclm = uwnd.sel(time=slice(start_date, end_date)).groupby("time.month").mean()
vclm = vwnd.sel(time=slice(start_date, end_date)).groupby("time.month").mean()
zclm = hgt.sel(time=slice(start_date, end_date)).groupby("time.month").mean()

# Calculate monthly anomalies
za = hgt.groupby("time.month") - zclm
za = za.sel(time=yyyymm).squeeze()

za.plot.contourf()
```

### Calculate WAF

i) Get grid delta and parameter somthing

```python
Ra = 287 # gas constant The ideal gas law is: pV = nRT, where n is the number of moles, and R is universal gas constant. 
         # The value of R depends on the units involved, but is usually stated with S.I. units as: R = 8.314 J/molK. 
         # This means that for air, you can use the value R = 287 J/kgK.
a = 6400000 # Earth's radius

lats = uclm.lat
lons = uclm.lon
lon, lat = np.meshgrid(lons, lats)

# 0.0872639
deg2rad = pi/180

dlon = np.gradient(lons)*deg2rad*2*units.rad
dlon = dlon[0]

dlat = np.gradient(lats)*deg2rad*2*units.rad
dlat = dlat[0]

coslat = np.cos(lat*deg2rad)*units.rad

#f = 2*7.24/100000*sinlat #/ units.rad * units['1/s']
f = mpcalc.coriolis_parameter(lat*units.degrees)
#g = 9.8 
```

ii) Calculate QG stream function

```python
psia=g/f*za
psia.plot.contourf()
psia
```

iii) Calculate magnitude of basic state wind speed

```python
magU = mpcalc.wind_speed(uclm.squeeze() * units['m/s'], 
                         vclm.squeeze() * units['m/s'])
magU.plot.contourf()
```

iv) Calculate first derivates

```python
#define dpsidlon = cdiff(psia,x)/dlon
dpsidlon = np.gradient(psia.values, dlon)[1]
ddpsidlonlon = np.gradient(dpsidlon, dlon)[1]

dpsidlat = np.gradient(psia.values, dlat)[0]
ddpsidlatlat = np.gradient(dpsidlat, dlat)[0]
ddpsidlatlon = np.gradient(dpsidlat, dlon)[1]

termxu = dpsidlon*dpsidlon-psia.values*ddpsidlonlon
termxv = dpsidlon*dpsidlat-psia.values*ddpsidlatlon
termyv = dpsidlat*dpsidlat-psia.values*ddpsidlatlat

fig, ax=plt.subplots(1,3, figsize=(12,3))

tmp = xr.DataArray(termxu, coords=za.coords)
tmp.plot.contourf(ax=ax[0])

tmp = xr.DataArray(termxv, coords=za.coords)
tmp.plot.contourf(ax=ax[1])

tmp = xr.DataArray(termyv, coords=za.coords)
tmp.plot.contourf(ax=ax[2])
```

Get px and py terms

```python
* "p" is normalized by 1000hPa
define coeff=coslat*(lev/1000)/(2*magU)
*x-component
*define px = coeff/(a*a*coslat)*( uclm*termxu+ vclm/coslat*termxv)
* corrected on 14th Sep. 2015: Thanks to Yan Jin.
define px = coeff/(a*a*coslat)*( uclm*termxu/coslat + vclm*termxv)

*y-component
define py = coeff/(a*a)*( uclm/coslat*termxv + vclm*termyv)

set lon 0 360

set gxout contour

* QG stream-function
set black -0.1 0.1
set cthick 20
set cint 5000000
d maskout( psia,  abs(lat)-10)

* horizontal wave-activity flux
set arrscl 0.5 15
set arrowhead -0.3
* maskout regions where 
* (i) QG approximation is supposed to be invalid (lower latitudes)
* (ii) westerly wind speed is weak or negative 
d skip(px,6,4);maskout(maskout( py , abs(lat)-10),magU-5)

draw title (a) WAF in Nov2014 Takaya&Nakamura 2001
printim Nov2014-TN01.png
```
